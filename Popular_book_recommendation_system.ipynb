# Collaborative based filtering

```python
x= rating_with_name.groupby('User-ID').count()['Book-Rating']>200
# This groups the data by userid and counts the number of book rating entries for each user. it then
# checks which users have rated more than 200 books, returning a boolean series. (true for users with more than 200 
# and false for others)
```

```python
smart_people =x[x].index
# x[x].index: This selects the True entries from x (i.e., users with more than 200 ratings) 
#     and retrieves their indices (which are the User-IDs in this case).
```

```python
smart_people
```

```python
filtered_rating= rating_with_name[rating_with_name['User-ID'].isin(smart_people)]
```

```python
filtered_rating
```

```python
y=filtered_rating.groupby('Book-Title').count()['Book-Rating']>=50
```

```python
y[y]
```

```python
famous_books =y[y].index
```

```python
final_ratings =filtered_rating[filtered_rating['Book-Title'].isin(famous_books)]
```

```python
final_ratings
```

```python
pt =final_ratings.pivot_table(index ='Book-Title',columns= 'User-ID', values ='Book-Rating')
```

```python
pt
```

```python
pt.fillna(0, inplace = True)
```

```python
pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
cosine_similarity(pt)
```
